**1. What is the difference between a function and a method in Python?**

Function:

- Defined using def (outside a class).

- Independent block of code.

- Called directly by name.

  Example:
```
def greet(name):
    return f"Hello {name}"
    greet("Alex")
```
Method:

- Defined inside a class.

- Associated with objects or the class.

- Called using dot notation (object.method()).

  Example:

```
class Person:
    def greet(self, name):
        return f"Hello {name}"
p = Person()
p.greet("Alex")
```
**2. Explain the concept of function arguments and parameters in Python.**

Parameters:

- Act as placeholders for values.

  Example:

Arguments:

- Actual values passed to a function when calling it.

  Example:

```greet("Alex")   # "Alex" is an argument```

Key Point: Parameters are what the function expects, arguments are what you give.

**5. What are iterators in Python and how do they differ from iterables?**

* **Iterable**

  * An object that can be looped over.
  * Examples: list, tuple, string, dict, set.
  * Provides an **iterator** when passed to 'iter()'.

  ```python
  nums = [1, 2, 3]     # iterable
  for n in nums:       # looping
      print(n)
  ```

* **Iterator**

  * An object with **state** that remembers where it is during iteration.
  * Has two main methods:

    * `__iter__()` → returns the iterator object
    * `__next__()` → returns the next item, raises `StopIteration` when finished

  ```python
  nums = [1, 2, 3]
  it = iter(nums)       # create iterator
  print(next(it))       # 1
  print(next(it))       # 2
  print(next(it))       # 3
  ```

**Key Difference:**

* Iterable = collection you can loop through.
* Iterator = the object that actually gives values one at a time.

**6. Explain the concept of generators in Python and how they are defined.**

  * Generators are a special type of iterator.
  * They generate values on the fly (one at a time) instead of storing them in memory.
  * Useful for handling large data efficiently.

* **How to Define**

  1. **Using 'yield' in a function**

     ```python
     def my_gen():
         yield 1
         yield 2
         yield 3
     for val in my_gen():
         print(val)
     ```

     'yield' pauses the function, saving its state, and resumes on the next call.

  2. **Generator Expression** (like list comprehension but with '()')

     ```python
     gen = (x*x for x in range(5))
     for val in gen:
         print(val)
     ```

**7. What are the advantages of using generators over regular functions?**

- **Memory Efficient** → Generate values one at a time (don’t store whole data in memory).

   ```python
   (x for x in range(1000000))   # uses less memory than a list
   ```

- **Lazy Evaluation** → Values are produced only when needed, not all at once.

- **Infinite Sequences** → Can represent endless streams of data.

   ```python
   def counter():
       n = 1
       while True:
           yield n
           n += 1
   ```

- **Improves Performance** → Faster for large datasets since it avoids creating huge lists.

- **Clean & Simple Code** → Easier to write with 'yield' than managing iterators manually.


**8. What is a lambda function in Python and when is it typically used?**

* **Definition**

  * A **lambda function** is a small, **anonymous (nameless) function** defined with the keyword `lambda`.
  * Syntax:
    ```python
    lambda arguments: expression
    ```
* **Example**

  ```python
  square = lambda x: x * x
  print(square(5))   # 25
  ```
* **Typical Uses**
  When a function is **simple** and used **temporarily**
  Often with functions like **`map()`**, **`filter()`**, **`sorted()`**

  ```python
  nums = [1, 2, 3, 4]
  doubled = list(map(lambda x: x*2, nums))
  print(doubled)   # [2, 4, 6, 8]
  ```

**9. Explain the purpose and usage of the 'map()' function in Python.**

* **Purpose**

  - Applies a **function** to each item of an iterable (like list, tuple).
  - Returns a **map object** (iterator), which can be converted to list/tuple.

* **Syntax**

  ```python
  map(function, iterable)
  ```

* **Example**

  ```python
  nums = [1, 2, 3, 4]
  squares = map(lambda x: x*x, nums)
  print(list(squares))   # [1, 4, 9, 16]
  ```

* **Usage**
  - When you want to **transform all items** in a sequence.
  - Often used with **lambda functions** for short operations.
  - Saves writing explicit loops.

**10. What is the difference between 'map()', 'reduce()', and 'filter()' functions in Python?**

| Function                            | Purpose                                                                  | Returns                         | Example                                       |
| ----------------------------------- | ------------------------------------------------------------------------ | ------------------------------- | --------------------------------------------- |
| **map()**                         | Applies a function to **each element** of an iterable                    | Iterator (transformed elements) | map(lambda x: x*2, [1,2,3]) → [2,4,6]       |
| **filter()**                      | Keeps elements that satisfy a **condition**                              | Iterator (filtered elements)    | filter(lambda x: x%2==0, [1,2,3,4]) → [2,4] |
| **reduce()** *(from functools)* | **Reduces** iterable to a single value by repeatedly applying a function | Single value                    | reduce(lambda x,y: x+y, [1,2,3,4]) → 10     |

---

### **Quick Examples**

```python
from functools import reduce

# map()
nums = [1, 2, 3, 4]
print(list(map(lambda x: x*2, nums)))   # [2, 4, 6, 8]

# filter()
print(list(filter(lambda x: x%2==0, nums)))   # [2, 4]

# reduce()
print(reduce(lambda x, y: x+y, nums))   # 10
```
**Key Point:**

* map() → transform all items
* filter() → select some items
* reduce() → combine into one value

````python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

my_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_evens(my_list))
````

````python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:  # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):  # check divisibility up to sqrt(n)
        if n % i == 0:
            return False
    return True


# Check from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is Prime")
    else:
        print(num, "is Not Prime")
````

````python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms   # total terms to generate
        self.count = 0
        self.a, self.b = 0, 1    # first two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
            raise StopIteration



fib = Fibonacci(10)   # generate first 10 terms
for num in fib:
    print(num, end=" ")
````

````python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(n):
    for exp in range(n + 1):
        yield 2 ** exp   # generate 2^exp



for val in powers_of_two(5):
    print(val, end=" ")
````

````python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

# Step 1: Create a file and write some lines into it
with open("sample.txt", "w") as f:
    f.write("Hello World\n")
    f.write("This is line 2\n")
    f.write("This is line 3\n")

# Step 2: Generator function to read the file line by line
def read_file_lines(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()   # strip() removes the newline character

# Step 3: Use the generator to read lines
for line in read_file_lines("sample.txt"):
    print(line)
````

````python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Convert to Fahrenheit using map + lambda
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print("Celsius:", celsius)
print("Fahrenheit:", fahrenheit)
````

````python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

def remove_vowels(text):
    vowels = "aeiouAEIOU"
    # filter keeps only those characters NOT in vowels
    return "".join(filter(lambda ch: ch not in vowels, text))


string = "Hello World, Python is Amazing!"
result = remove_vowels(string)

print("Original:", string)
print("Without vowels:", result)
````
